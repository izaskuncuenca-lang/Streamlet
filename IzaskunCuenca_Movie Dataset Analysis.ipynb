```python
df['Year'] = df['title'].str.extract(r'\((\d{4})\)$')
```

```python
df['Title'] = df['title'].str.replace(r'\s*\(\d{4}\)$', '', regex=True).str.strip()
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce') # convierte el año de texto a número
```

```python
df = df.drop(columns=['title']) #elimino la columna del titulo porque ya no es necesario 
df = df[['movieId', 'Title', 'Year', 'genres']] #aqui las reordeno para que quede más limpio
```

```python
genres_exploded = df['genres'].str.split('|').explode() # aquío convierto los géneros en una lista y el explode crea una fila separada por cada género
```

```python
genre_counts = genres_exploded.value_counts() #aqui veo cuántas veces aparece cada género en todo el dataset
```

```python
genre_counts_filtered = genre_counts[genre_counts.index != '(no genres listed)'] #Filtro el valor (no genres listed) porque no es un género real y no quiero que aparezca en las gráficas
```

```python
movies_per_year = df.dropna(subset=['Year']).groupby('Year').size() # el dropna year quita las filas donde no hay año, 
#el groupby year agrupa todas las películas por año y cuenta cuántas hay en cada uno 
```

```python
df['Decade'] = (df['Year'] // 10 * 10).astype('Int64') #convierte el año en década, mas facil de agruparlas
```

```python
df['num_genres'] = df['genres'].str.split('|').apply(len) #Separamos los géneros por | y con len contamos cuántos géneros tiene cada película
# esto nos permite ver si las películas tienden a ser de 1 género, 2, 3, o mas
```
